```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy.fft import fft, fftfreq, fftshift
from scipy.signal import get_window
import random
```

```python
# Generate 100 sine functions with different a,b,c parameters and random noise
def sine(a,b,c,x):
    return a * np.sin(2*np.pi*b * x + c)

x_data = np.arange(0,10,0.001)
a_list = np.linspace(5,10,100)
b_list = np.linspace(0.1,3,100)
c_list = np.linspace(0,5,100)

my_a = random.choice(a_list)
my_b = random.choice(b_list)
my_c = random.choice(c_list)

y_data = sine(my_a, my_b, my_c, x_data) #+ my_a * np.random.rand(len(x_data))
y_err = np.random.rand(len(x_data))
```

```python
plt.errorbar(x_data, y_data, y_err)
#plt.xlim(0,1)
#plt.ylim(0,1)
```

```python
def fft(x,y):
    N = len(y)
    dx = (x[1]-x[0])
    fs = int(1.0 / dx)
    window = get_window('hann', y.size)
    win_data = y * window
    freq = np.fft.rfftfreq(len(win_data))*fs
    freq_domain = np.fft.rfft(win_data)/fs
    return freq, abs(freq_domain)

freqs, amps = fft(x_data, y_data)
plt.plot(freqs, amps)
plt.xlim(0,10)

for i in range(0,len(amps)):
    if amps[i] == np.max(amps):
        freq = freqs[i]
print('FFT freq: ',freq)
print('b: ',my_b)
```

```python
index = np.argwhere(x_data <= round(1/freq, 3)).flatten()[-1:]
x_data = x_data[0:index[0]]
y_data = y_data[0:index[0]]
y_err = y_err[0:index[0]]
plt.errorbar(x_data, y_data, y_err)
```

I copied the code below from the Intro to Regression notebook:

```python
def lnlikelihood(a, b, c, x_data, y_data, y_err):
    model = sine(a,b,c,x_data)
    lnlike = -.5*np.sum(np.log(2*np.pi*y_err**2)) - np.sum((y_data - model)**2/(2*y_err**2)) 
    return lnlike
```

Let's define an initial m and b, sigma for the normal distribution of m and b, the number of steps in the chain, and arrays to store all the m and b values: 

```python
a_current = 200
b_current = freq
c_current = 200
sigma_a = .1
sigma_b = .1
sigma_c = .1
N = 10

a_chain = []
b_chain = []
c_chain = []
```

Let's just say the prior is 1, so the posterior equals the likelihood. The for loop below generates a new m and b from a normal distribution centered around the current m and b, then calculates the posterior for the current m and b and the new m and b, then decides whether to accept or reject the new m and b based on whether the ratio of the posteriors is above a randomly generated number. If the new values are accepted, m and b are updated to the new m and b.

```python
for _ in range(N):
    a_new = np.random.normal(loc=a_current, scale=sigma_a)
    b_new = np.random.normal(loc=b_current, scale=sigma_b)
    c_new = np.random.normal(loc=c_current, scale=sigma_c)

    posterior_current = lnlikelihood(a_current, b_current, c_current, x_data, y_data, y_err)
    posterior_new = lnlikelihood(a_new, b_new, c_new, x_data, y_data, y_err)

    ratio = np.exp(posterior_new)/np.exp(posterior_current)
    print(posterior_new-posterior_current)

    accept_reject = np.random.uniform(low=0., high=1.)

    if ratio > accept_reject:
        a_current = a_new
        b_current = b_new
        c_current = c_new
        a_chain.append(a_current)
        b_chain.append(b_current)
        c_chain.append(c_current)
    else:
        a_chain.append(a_current)
        b_chain.append(b_current)
        c_chain.append(c_current)
```

As we saw in the Intro to Regression notebook, a slope of about 1.08 and intercept of about 217 is an alright fit for the data, so let's see if our m and b settle around those values.

```python
step = np.arange(1,N+1,1)
plt.plot(step,a_chain,label='chain')
plt.plot(step, [my_a]*len(step),label='actual')
plt.legend()
plt.xlabel('step')
plt.ylabel('a')
```

```python
plt.plot(step,b_chain, label='chain')
plt.plot(step, [my_b]*len(step),label='actual')
plt.legend()
plt.xlabel('step')
plt.ylabel('b')
```

```python
plt.plot(step,c_chain, label='chain')
plt.plot(step, [my_c]*len(step), label='actual')
plt.legend()
plt.xlabel('step')
plt.ylabel('c')
```

Now let's set up the initial m and b. I picked these values based on the line I used to cut the white dwarfs, since that line looks like about the right slope, but I wanted the initial slope and intercept guess to be off by a little so I can see the burn-in to confirm the MCMC is working like it should.

```python
m_current = 2
b_current = 10
_current = 2

sigma_m = 1
sigma_b = 1
sigma_ = 1

N = 10000

ms = [m_current]
bs = [b_current]
s = [_current]
```

Now we can redo the Markov chain for loop from earlier, but this time the posterior equals the prior times the likelihood. To keep things in log form as much as possible, I made an if statement to make sure the prior is always satisfied for both the current $\sigma _{M_G}$ guess and the new $\sigma_{M_G}$ guess.

```python
for _ in range(N):
    m_new = np.random.normal(loc=m_current, scale=sigma_m)
    b_new = np.random.normal(loc=b_current, scale=sigma_b)
    _new = np.random.normal(loc=_current, scale=sigma_)

    if prior(_current) == 1 and prior(_new) == 1:
        posterior_current = lnlikelihd(m_current, b_current, _current)
        posterior_new = lnlikelihd(m_new, b_new, _new)
        ratio = np.exp(posterior_new - posterior_current)
    else:
        ratio = 0

    accept_reject = np.random.uniform(low=0., high=1.)

    if ratio > accept_reject:
        m_current = m_new
        b_current = b_new
        _current = _new
        ms.append(m_current)
        bs.append(b_current)
        s.append(_current)
    else:
        ms.append(m_current)
        bs.append(b_current)
        s.append(_current)
```
